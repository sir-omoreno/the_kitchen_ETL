```python
import pandas as pd
import requests
import json
from config import api_id_ru, api_key_ru
```

```python
 # Popular list of ingredients from Healthline website
pop_list = ['eggs', 'almonds', 'chicken breast', 'oats', 'cottage cheese', 'greek yogurt', 'milk', 'broccoli',
                'lean beef','tuna','quinoa', 'whey protein supplements', 'lentils','ezekiel bread','pumpkin seeds',
                'turkey breast','fish', 'shrimp', 'brussel sprouts','peanuts']
#short_pop_list = ['eggs', 'almonds', 'chicken breast', 'oats']
#for ingredient in short_pop_list:
ingredient = 'greek yogurt'
# Cooking time greater than 5 minutes
time = '5%2B'
# Pull in information for top 50 recipes for the "ingredient"
from_number = "0"
to_number = "50"
# Url for Edamam Api
url = "https://api.edamam.com/search" + "?q=" + ingredient + "&app_id=" + api_id_ru + "&app_key=" + api_key_ru + "&time=" + time + "&from=" + from_number + "&to=" + to_number
print(url)
```

```python
food_request = requests.get(url)
food_json = food_request.json()
print(food_json)
```
