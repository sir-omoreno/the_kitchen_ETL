```python
import pandas as pd
from sqlalchemy import create_engine
from config import password
```

```python
file = "datasets/data.csv"
recipes_df = pd.read_csv(file)
```

```python
recipes_df
```

```python
connection = f"postgres:{password}@localhost:5432/ETL"
engine = create_engine(f'postgresql://{connection}')
```

```python
engine.table_names()
```

```python
recipetable_df = recipes_df[["Main Ingredient","Recipe Name","Cook Time","Yield","Calories"]]
recipetable_df.rename(columns={"Main Ingredient":"Main_Ingredient","Recipe Name":"Recipe_Name","Cook Time":"Cook_time"},inplace=True)

recipetable_df.index.name = "id"
recipetable_df
```

```python
urltable_df = recipes_df[["Main Ingredient", "URL"]]
urltable_df.rename(columns={"Main Ingredient":"Main_Ingredient"},inplace=True)
urltable_df.index.name = "id"
urltable_df
```

```python
# Ingredient Table
ingredient_list = recipes_df["Main Ingredient"].unique()
ingredient_df = pd.DataFrame(ingredient_list, columns=['Main_Ingredient'])
ingredient_df.index.name = "id"
ingredient_df.index += 1 
ingredient_df
```

```python
# Load dataframes into database, Table: Recipes
recipetable_df.to_sql(name='Recipes', con=engine, if_exists='append', index=True)
```

```python
# Load dataframes into database, Table: Url
urltable_df.to_sql(name='Url', con=engine, if_exists='append', index=True)
```

```python
# Load dataframes into database, Table: Ingredients
ingredient_df.to_sql(name='Ingredients', con=engine, if_exists='append', index=True)
```
